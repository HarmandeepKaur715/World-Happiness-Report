```python
# Single Country
# To plot a country's happiness trend across years, use df_combined and filter for the specific country.
# For example, to plot the trend for Finland (which is the first country in df):
country_to_plot = df['Country_or_region'].iloc[0] # Get the first country from the current df
country_data_combined = df_combined[df_combined['Country'] == country_to_plot]

plt.plot(country_data_combined['Year'], country_data_combined['Score'], marker='s', color = "red")
plt.title(f'{country_to_plot} Happiness Trend')
plt.xlabel('Year')
plt.ylabel('Score')
plt.show()
```

```python
#Multiple couontries
countries = df['Country_or_region'].unique()[:5]

for c in countries:
    data = df_combined[df_combined['Country'] == c] # Use df_combined and the 'Country' column
    plt.plot(data['Year'], data['Score'], marker='o', label=c)

plt.legend()
plt.title('Country Comparison')
plt.xlabel('Year')
plt.ylabel('Score')
plt.show()
```
